```python
# Install required packages
!pip install datasets evaluate pandas tqdm

# Mount Google Drive (if needed)
from google.colab import drive
drive.mount('/content/drive')

# Import necessary modules
import os
import json
import pandas as pd
from datasets import load_dataset
import evaluate
from tqdm import tqdm

# Language mappings
FLORES_MAP = {
    "Assamese": "asm_Beng",
    "Bengali": "ben_Beng",
    "Gujarati": "guj_Gujr",
    "Hindi": "hin_Deva",
    "Kannada": "kan_Knda",
    "Malayalam": "mal_Mlym",
    "Marathi": "mar_Deva",
    "Nepali": "npi_Deva",
    "Odia": "ory_Orya",
    "Punjabi": "pan_Guru",
    "Sanskrit": "san_Deva",
    "Tamil": "tam_Taml",
    "Telugu": "tel_Telu",
    "Urdu": "urd_Arab",
    "Bodo": "brx_Deva",
    "Dogri": "doi_Deva",
    "Konkani": "gom_Deva",
    "Kashmiri": "kas_Arab",
    "Maithili": "mai_Deva",
    "Manipuri": "mni_Mtei",
    "Santali": "sat_Olck",
    "Sindhi": "snd_Deva"
}

LANG_MAP = {
    "asm": "Assamese",
    "ben": "Bengali",
    "guj": "Gujarati",
    "hin": "Hindi",
    "kan": "Kannada",
    "mal": "Malayalam",
    "mar": "Marathi",
    "npi": "Nepali",
    "ory": "Odia",
    "pan": "Punjabi",
    "san": "Sanskrit",
    "tam": "Tamil",
    "tel": "Telugu",
    "urd": "Urdu",
    "brx": "Bodo",
    "doi": "Dogri",
    "gom": "Konkani",
    "kas": "Kashmiri",
    "mai": "Maithili",
    "mni": "Manipuri",
    "sat": "Santali",
    "snd": "Sindhi"
}

def compute_bleu_score(translations_dir, lang_code):
    """
    Computes the BLEU score for a given language while handling missing translations.

    Args:
        translations_dir (str): Path to the directory containing translation JSONL files.
        lang_code (str): 3-letter language code (e.g., 'mar', 'ben').
    """
    bleu = evaluate.load("bleu")
    print(f"\nEvaluating BLEU score for language: {lang_code} ({LANG_MAP.get(lang_code, 'Unknown')})")

    # Collect all JSONL files for the current language
    lang_files = [f for f in os.listdir(translations_dir) if f.startswith(f"{lang_code}_translations_") and f.endswith('.jsonl')]

    if not lang_files:
        print(f"No translation files found for language code: {lang_code}")
        return

    translations = {}

    for json_file in lang_files:
        file_path = os.path.join(translations_dir, json_file)
        with open(file_path, 'r', encoding='utf-8') as f:
            try:
                data = json.load(f)
                translation = data.get('translated_text', '').strip()
                idx = data.get('id')
                if translation and idx is not None:
                    translations[idx] = translation
            except json.JSONDecodeError:
                print(f"Error decoding JSON from file: {json_file}")

    if not translations:
        print(f"No valid translations found in files for language code: {lang_code}")
        return

    # Load reference dataset
    try:
        # Correctly construct the configuration name without 'sentence_' prefix
        config = f"eng_Latn-{FLORES_MAP[LANG_MAP[lang_code]]}"
        dataset = load_dataset("ai4bharat/IN22-Conv", config, trust_remote_code=True)
        df = pd.DataFrame(dataset["conv"])

        # Construct the reference column name by adding 'sentence_' prefix
        ref_column = f"sentence_{FLORES_MAP[LANG_MAP[lang_code]]}"
        if ref_column not in df.columns:
            print(f"Reference column '{ref_column}' not found for language code: {lang_code}.")
            print(f"Available columns: {df.columns.tolist()}")
            return
    except Exception as e:
        print(f"Error loading reference dataset for {lang_code}: {e}")
        return

    # Filter references based on IDs available in translations
    matching_ids = set(translations.keys())
    references_df = df[df['id'].isin(matching_ids)].sort_values('id')
    references = references_df[ref_column].tolist()
    reference_ids = references_df['id'].tolist()

    # Align translations with references by matching IDs
    aligned_translations = [translations[id] for id in reference_ids if id in translations]

    # Check alignment
    if len(aligned_translations) != len(references):
        min_len = min(len(aligned_translations), len(references))
        aligned_translations = aligned_translations[:min_len]
        references = references[:min_len]
        print(f"Adjusted number of translations and references to {min_len} for language code: {lang_code}")

    # Prepare references for BLEU
    references = [[ref] for ref in references]

    # Compute BLEU
    results = bleu.compute(predictions=aligned_translations, references=references)
    print(f"BLEU score for {lang_code} ({LANG_MAP[lang_code]}): {results['bleu']*100:.2f}")

def main():
    # Path to your translations directory in Google Drive
    translations_dir = r"/content/drive/MyDrive/all"  # Update this path to your translations directory

    # List of all 22 Indic language codes (3-letter)
    language_codes = [
        "asm", "ben", "guj", "hin", "kan", "mal", "mar",
        "npi", "ory", "pan", "san", "tam", "tel", "urd",
        "brx", "doi", "gom", "kas", "mai", "mni", "sat", "snd"
    ]

    for lang in language_codes:
        compute_bleu_score(translations_dir, lang)

if __name__ == "__main__":
    main()
```
